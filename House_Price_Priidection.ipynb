# House Price Prediction Based on Area

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

# DataSet 

```python

# Create the DataFrame
df = {
    "Area": [2600, 3000, 3200, 3600, 4000],
    "Price": [550000, 565000, 610000, 680000, 725000]
}
data = pd.DataFrame(df)

# Display the DataFrame
print(data)
data.to_csv("PricePrediction.csv")
# Plotting
%matplotlib inline
plt.xlabel("Area")
plt.ylabel("Price")
plt.scatter(data["Area"], data["Price"], color="red", marker="+")
plt.show()
```

# Fit linear Regrression Model

```python
model=linear_model.LinearRegression()
```

```python
df_new=data.drop("Price",axis="columns")
```

```python
df_new
```

```python
model.fit(df_new,data.Price)
```

# Prediction Test 

```python
model.predict([[5000]])
```
